```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import r2_score

file_name = "choclate protfolio project - 11.csv"
df = pd.read_csv(file_name)
print("1. Dataset successfully loaded! Shape:", df.shape)


def clean_currency(val):
    if isinstance(val, str):
       
        val = val.replace('$', '').replace(',', '').replace('%', '').strip()
        
      
        if '(' in val and ')' in val:
            val = '-' + val.replace('(', '').replace(')', '')
            
        try:
            return float(val)
        except ValueError:
            
            return 0.0
    return val


for col in ['Amount', 'Cost', 'Profit', 'profit %']:
    df[col] = df[col].apply(clean_currency)

print("2. Data cleaning completed! (Excel errors like #DIV/0! fixed successfully)")


X = df[['Sales Person', 'Geography', 'Product', 'cost per unit']]
y = df['Units']


categorical_cols = ['Sales Person', 'Geography', 'Product']
numerical_cols = ['cost per unit']


preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ])

X_processed = preprocessor.fit_transform(X).toarray()


X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)
print("3. Preprocessing done. Training samples:", X_train.shape[0], "| Test samples:", X_test.shape[0])


input_dim = X_train.shape[1]

model = Sequential([
    Dense(64, activation='relu', input_shape=(input_dim,)),  
    Dropout(0.2),                                            
    Dense(32, activation='relu'),                           
    Dense(1, activation='linear')                            
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()


print("\n--- 4. ANN Model Training Shuru Ho Rahi Hai ---")
history = model.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=16, verbose=1)


y_pred = model.predict(X_test).flatten()


rmse = np.sqrt(np.mean((y_test - y_pred)**2))
r2 = r2_score(y_test, y_pred)

print(f"\n=======================================")
print(f"       FINAL MODEL EVALUATION         ")
print(f"=======================================")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R2) Score: {r2:.4f}")
print(f"=======================================")

plt.figure(figsize=(8, 5))
plt.plot(history.history['loss'], label='Train Loss (MSE)')
plt.plot(history.history['val_loss'], label='Validation Loss (MSE)')
plt.title('ANN Model Training History')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.grid(True)


plt.savefig('loss_chart.png')
print("\n5. Loss progress chart successfully saved as 'loss_chart.png'")
```
